```python
#@title  Puxar Base de Dados

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, date
from sklearn import preprocessing
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

#Importando bibliotecas necessárias

uri = '/content/drive/Shareddrives/Colab_Grupo2/Colab/Data/dataToPredictTarget.csv'

data = pd.read_csv(uri)

#Declarando caminho do conjunto de dados e instanciando um dataframe
```

```python
#@title Algoritmo de instanciamento de vetores

#@markdown Execute o algoritmo para dividir as colunas numéricas e classifivativas em vetores

numeric_values =["bmi", "weight", "height", "partos", "breast_feeding_time", "menarche", "tempo_rep_hormo", "sister_n", "daughter_n", "birads", "nuclear_grade", "mitotic_index", "grau_hist", "histological_grade", "progesteron_perct", "er_percentage", "ki67_perc", "follow_up_days", "follow_up_days_recidive", "age", "treatment_time", "exam_time"]

class_values = ["t", "n", "tumor_stage", "m", "topo", "morfo","trathosp"]

data_classified = data.copy()
```

```python
# #@title Algoritmo de agrupamento

# #@markdown Execute o algoritmo para agrupar conjunto de valores numéricos

for z in numeric_values:
  if z in data.columns:
    for i in range(data[z].size):
      if data.iloc[i][z] <= data[z].quantile(0.25):
        value = 1
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
      elif data.iloc[i][z] <= data[z].quantile(0.5):
        value = 2
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
      elif data.iloc[i][z] <= data[z].quantile(0.75):
        value = 3
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
      else:
        value = 4
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
    data_classified[z] = data_classified[z].astype('int')


```

```python
#@title Algoritmo de agrupamento

#@markdown Execute o algoritmo para agrupar conjunto de valores numéricos

for z in numeric_values:
  if z in data.columns:
    for i in range(data[z].size):
      if data.iloc[i][z] <= data[z].quantile(0.25):
        value = 0
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
      elif data.iloc[i][z] <= data[z].quantile(0.5):
        value = 1
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
      elif data.iloc[i][z] <= data[z].quantile(0.75):
        value = 2
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
      else:
        value = 3
        data_classified[z].replace(data.iloc[i][z], str(value), inplace=True)
    data_classified[z] = data_classified[z].astype('int')
```

```python
for z in class_values:
  if z in data.columns:
    value = 1
    sorted = data[z].unique()
    sorted.sort()
    for i in sorted:
      data_classified[z].replace(i, str(value), inplace=True)
      value += 1
    data_classified[z] = data_classified[z].astype('int')

data_classified["escolari"].replace(9, 0, inplace=True)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore')

for i in class_values:
  x = pd.get_dummies(data_classified[i], prefix=[i])
  del data_classified[i]
  data_classified = pd.merge(data_classified, x, left_index=True, right_index=True)


```
