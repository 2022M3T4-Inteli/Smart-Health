#**Formulário para o usuário realizar a predição**


```python
#@title Number fields
bmi = 19.5 #@param {type:"number"}
weight = 0.122827 #@param {type:"number"}
height = 1.65 #@param {type:"number"}
antec_fam_cancer_mama = 2 #@param {type:"integer"}
alcohol_type___1 = 0 #@param {type:"number"}
er_ihc = 1 #@param {type:"number"}
pr_ihc = 1 #@param {type:"number"}
tumor_subtype = 2 #@param {type:"number"}
reclocal = 1 #@param {type:"number"}
recregio = 1 #@param {type:"number"}
recdist = 0 #@param {type:"number"}
trathosp = 5 #@param {type:"number"} 
morfo = 84803 #@param {type:"number"}
tumor_stage = 22 #@param {type:"number"}
t = 2 #@param {type:"number"}
n = 1 #@param {type:"number"}
m = 0 #@param {type:"number"}
exam_time_meses = 46 #@param {type:"number"}
her2_ihc = 1 #@param {type:"number"}
recidive = 1 #@param {type:"number"}
age = 45 #@param {type:"number"}
```

#**Configurações do formulário**

```python
#@title Armazenar os valores do formulário
values = {
    'bmi': bmi,
    'weight': weight,
    'height': height * 100,
    'antec_fam_cancer_mama': antec_fam_cancer_mama,
    'alcohol_type___1': alcohol_type___1,
    'er_ihc': er_ihc,
    'pr_ihc': pr_ihc,
    'tumor_subtype': tumor_subtype,
    'reclocal': reclocal,
    'recregio': recregio,
    'recdist': recdist,
    'trathosp': trathosp,
    'morfo': morfo,
    'tumor_stage': tumor_stage,
    't': t,
    'n': n,
    'm': m,
    'exam_time': exam_time_meses * 30,
    'her2_ihc': her2_ihc,
    'recidive': recidive,
    'age': age * 365
}
```

```python
#@title Salvar os valores do formulário em um dataframe

df = pd.DataFrame([values])
```

```python
df.head()
```

```python
#@title Definir resultados

result = model.predict(df)[0]
if result == 0:
  saida = ("baixa")
elif result == 1:
  saida = ("media")
else:
  saida = ("alta")
```
